Tijdsintervallen bij een gooi bepalen

```python
def gooi_checker(data, N, gamma):
    luchttijd = []
    gooitijd = []
    
    # Oprapen na stilstand op grond
    verg_data = data.copy() - data.iloc[0]
    gooi_data = data.copy()
    gooi_data['squared norm'] = verg_data['x_acc'] ** 2 + verg_data['y_acc'] ** 2 + verg_data['z_acc'] ** 2

    # Zero acceleration na loslaten
    za_data = data.copy()
    za_data['squared norm'] = data['x_acc'] ** 2 + data['y_acc'] ** 2 + data['z_acc'] ** 2

    done = False

    total_lucht_duration = 0
    total_gooi_duration = 0
    total_raap_duration = 0

    while not done:
        gedaan = False
        voltooid = False
        luchttijd.clear()
        gooitijd.clear()

        for i in range(N, len(za_data) - N):
            # Herken wanneer dobbelsteen wanneer wordt opgepakt
            if all(waarde > 0.01 * gamma for waarde in gooi_data['squared norm'][i-N:i+N]) and not gedaan:
                gooitijd.append(data['timestamp'][i])
                gedaan = True
                
            # Herken wanneer dobbelsteen de hand verlaat
            elif all(element < gamma for element in za_data['squared norm'][i-N:i+N]) and not voltooid:
                temp_luchttijd = list(data['timestamp'][i-N:i+N])

                # Randvoorwaarde dat de dobbelsteen minstens 0.5 secondes wordt geschud
                if gooitijd and (temp_luchttijd[0] - gooitijd[-1]) > 500:
                    luchttijd.extend(temp_luchttijd)
                    gooitijd = sorted(gooitijd)
                    luchttijd = sorted(luchttijd)
                    
                    if gooitijd:
                        total_raap_duration = gooitijd[-1] - 2*N
                        
                    if luchttijd:
                        land_tijd = luchttijd[-1]
                        los_tijd = luchttijd[0] 
                        total_lucht_duration = land_tijd - los_tijd
                        
                    if gooitijd and luchttijd:
                        total_gooi_duration = los_tijd - total_raap_duration
                    done = True
                    break
                    
        if not done:
            gamma += 0.001

    total_starttotlos = total_raap_duration + total_gooi_duration  
    total_starttotgrond = total_lucht_duration + total_raap_duration + total_gooi_duration
    
    print("Start tot oprapen = ", total_raap_duration)
    print("Oprapen tot loslaten = ", total_gooi_duration)
    print("Loslaten tot grond = ", total_lucht_duration)
    print("Start tot loslaten = ", total_starttotlos)
    print("Start tot grond = ", total_starttotgrond)

    return total_raap_duration, total_gooi_duration, total_lucht_duration, total_starttotlos, total_starttotgrond

def side_detector(ya):
    sides = np.array([
        [0,0,1],
        [1,0,0],
        [0,1,0],
        [0,-1,0],
        [-1,0,0],
        [0,0,-1]
    ])
    norms = []
    for side in sides:
        norms.append(np.linalg.norm(side-ya))
    return np.argmin(np.array(norms)) + 1
```

```python
dob.connect()
dob.log(3, 400, 8, 2000)
dob.download()
data = dob.datadf

# Deze code niet vaker dan 1 keer uitvoeren!
data = remove_nan(data)
data = rotate_remove_bias(data, bias_cali, gyro_rot_quat)
```

```python
total_raap_duration, total_gooi_duration, total_lucht_duration, total_starttotlos, total_starttotgrond= gooi_checker(data, 11, 0.001)
ya = np.mean(data['x_acc'][-10:-1]), np.mean(data['y_acc'][-10:-1]), np.mean(data['z_acc'][-10:-1])
print("Aantal ogen boven = ", side_detector(ya))
```

```python
df = pd.DataFrame(data)

# Plot de accelerometer waardes
columns_to_plot = [col for col in df.columns if 'acc' in col]
df.plot(x='timestamp', y=columns_to_plot, kind='line')

# Vertical gray line for when the dice is picked up
plt.axvline(x=total_raap_duration, color='gray', linestyle='--', alpha=0.5, label='Start tot rapn')

# Vertical red line for when the dice leaves the hand
plt.axvline(x=total_starttotlos, color='gray', linestyle='--', alpha=0.5, label='Start tot loslaten')

# Vertical blue line for when the dice hits the ground
plt.axvline(x=total_starttotgrond, color='gray', linestyle='--', alpha=0.5, label='Start to eerste grond contact')


plt.xlabel('Timestamp')
plt.ylabel('Acceleration')
plt.title('Acceleration Data')
plt.xticks(range(int(df['timestamp'].min()), int(df['timestamp'].max()) + 500, 500))
plt.show()
```
