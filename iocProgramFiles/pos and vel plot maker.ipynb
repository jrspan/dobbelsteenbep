```python
import matplotlib.pyplot as plt

from RW_helper import *
import pandas as pd
```

```python
df = pd.read_csv('2024-06-13_13-13-54.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'px'], label='x position')
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'py'], label='y position')
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'pz'], label='z position')
plt.title('Position Estimate')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.grid()
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'vx'], label='x velocity')
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'vy'], label='y velocity')
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'vz'], label='z velocity')
plt.title('Velocity Estimate')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.grid()
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'x_acc'], label='x acceleration')
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'y_acc'], label='y acceleration')
plt.plot(df.loc[:, 'timestamp'] / 1000, df.loc[:, 'z_acc'], label='z acceleration')
plt.title('Raw Acceleration')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (g)')
plt.grid()
plt.legend()

plt.tight_layout()

plt.show()
```
